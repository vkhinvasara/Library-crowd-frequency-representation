```python
import matplotlib.pyplot as plt
import pandas as pd
```

Read Total profit of all months and show it using a line plot

```python
df = pd.read_csv('company_sales_data.csv')
profit = df['total_profit']
months = df['month_number']

plt.plot(months, profit)
plt.show()
```

Read all product sales data and show it using a multiline plot. Display the number of units
sold per month for each product using multiline plots. (i.e., Separate Plotline for each product ).


```python
facecream = df['facecream']
facewash = df['facewash']
toothpaste = df['toothpaste']
bathingsoap = df['bathingsoap']
shampoo = df['shampoo']
moisturizer = df['moisturizer']

plt.plot(months,facecream)
plt.plot(months,facewash)
plt.plot(months, toothpaste)
plt.plot(months, bathingsoap)
plt.plot(months, shampoo)
plt.plot(months, moisturizer)
plt.legend(["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"],loc = "upper left")

plt.show()
```

Read toothpaste sales data of each month and show it using a scatter plot

```python
plt.scatter(months, toothpaste)
plt.show()
```

Read face cream and facewash product sales data and show it using the bar chart

```python
plt.bar(months +0.2,facecream,0.4)
plt.bar(months -0.2 , facewash,0.4)
plt.legend(["facecream", "facewash"], loc = "upper right")
plt.show()
```

Calculate total sale data for last year for each product and show it using a Pie chart.

```python
label = ["facecream","facewash","toothpaste","bathingsoap","shampoo","moisturizer"]
total_profit = []
for columns in df:
    total_profit.append(sum(df[columns]))

total_profit.pop(0)
total_profit.pop()
total_profit.pop()
# print(total_profit)
plt.pie(total_profit, labels=label, autopct='%1.1f%%')
plt.show()
```
